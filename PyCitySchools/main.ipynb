```python
#Dependencies
import pandas as pd
from pathlib import Path
```

```python
#load file 
school_data_load = Path("Resources/schools_complete.csv")
student_data_load = Path("Resources/students_complete.csv")

#read schools data and store into Panda Dataframes 
school_data=pd.read_csv(school_data_load)
school_data.head()
```

```python
#read students data and store into Panda Dataframes 
student_data=pd.read_csv(student_data_load)
student_data.head()
```

Math Scores by Grades

```python
ninth_graders = school_students_data_complete[(school_students_data_complete["grade"] == "9th")]
tenth_graders =school_students_data_complete[(school_students_data_complete["grade"] == "10th")]
eleventh_graders = school_students_data_complete[(school_students_data_complete["grade"] == "11th")]
twelfth_graders = school_students_data_complete[(school_students_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = school_students_data_complete.loc[school_students_data_complete["grade"]=="9th"].groupby("school_name")["math_score"].mean()
tenth_grader_math_scores = school_students_data_complete.loc[school_students_data_complete["grade"]=="10th"].groupby("school_name")["math_score"].mean()
eleventh_grader_math_scores = school_students_data_complete.loc[school_students_data_complete["grade"]=="11th"].groupby("school_name")["math_score"].mean()
twelfth_grader_math_scores = school_students_data_complete.loc[school_students_data_complete["grade"]=="12th"].groupby("school_name")["math_score"].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    "9th":ninth_grade_math_scores,
    "10th":tenth_grader_math_scores,
    "11th":eleventh_grader_math_scores,
    "12th":twelfth_grader_math_scores
})

# Minor data wrangling
math_scores_by_grade = math_scores_by_grade[["9th", "10th", "11th", "12th"]]
math_scores_by_grade.index.name="school_name"

# Display the DataFrame
math_scores_by_grade
```

Scores by school spending

```python
school_summary.head()
```

```python
# Establish the bins.
bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.


school_summary['Total School Budget'] = school_summary['Total School Budget'].str.replace(',', '').str.replace('$', '')
school_summary['Total School Budget'] = pd.to_numeric(school_summary['Total School Budget'])
school_summary['spending_bins']= pd.cut(school_summary['Total School Budget']/school_summary['Total Students'], bins, labels=labels)




#group by spending
spending_bins = school_summary.groupby('spending_bins')
# included spendoing in school summary
school_summary.head()
```

```python



# calculations

size_math_scores =school_summary.groupby(["spending_bins"])["Average Math Score"].mean()
size_reading_scores = school_summary.groupby(["spending_bins"])["Average Reading Score"].mean()
size_passing_math = school_summary.groupby(["spending_bins"])["% Passing Math"].mean()
size_passing_reading = school_summary.groupby(["spending_bins"])["%Passing Reading"].mean()
size_overall_passing = school_summary.groupby(["spending_bins"])["Overall Passing rate"].mean()


#the new dataframe called school_spending_df

school_spending_df = pd.DataFrame({
"Average Math Scores":size_math_scores,
"Average Reading Scores":size_reading_scores,
"% Passing Math":size_passing_math,
"% Passing Reading":size_passing_reading,
"Overall Passing":size_overall_passing

})
#index
reading_scores_by_grade.index.name="spending_bins"

#display dataframe
school_spending_df
```
